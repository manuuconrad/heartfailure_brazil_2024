```python
import pandas as pd

# 1. Carregue os dois (como você limpou no Excel, talvez não precise do skiprows)
df_int = pd.read_csv('internacoes_ic.csv', sep=';', encoding='iso-8859-1')
df_obt = pd.read_csv('obitos_ic.csv', sep=';', encoding='iso-8859-1')

# 2. O comando mágico para unir tabelas baseando-se em uma coluna igual:
# Vamos unir pela coluna do Estado (certifique-se que o nome é igual nas duas)
df_final = pd.merge(df_int, df_obt, on='Região/Unidade da Federação', suffixes=('_int', '_obt'))

# 3. Dê um head para ver como ficou
df_final.head()
```

```python
# 1. Limpeza de segurança: remove espaços invisíveis que o Excel adora criar
df_final.columns = df_final.columns.str.strip()

# 2. Lista com os nomes EXATOS das suas colunas (ajuste se precisar)
meses_colunas = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

for mes in meses_colunas:
    # Montando os nomes com os sufixos que o merge criou
    col_int = f'2024/{mes}_int'
    col_obt = f'2024/{mes}_obt'
    col_res = f'Letalidade_{mes}'

    # Verificação: Só faz a conta se as duas colunas existirem no seu df_final
    if col_int in df_final.columns and col_obt in df_final.columns:
        # Converte para número (forçando a barra se for preciso)
        int_num = pd.to_numeric(df_final[col_int].astype(str).str.replace('.', '', regex=False), errors='coerce')
        obt_num = pd.to_numeric(df_final[col_obt].astype(str).str.replace('.', '', regex=False), errors='coerce')

        # Faz a conta da letalidade
        df_final[col_res] = (obt_num / int_num) * 100
    else:
        print(f"Atenção: Não achei as colunas para o mês {mes}")

# 3. Vendo o resultado
print("Cálculo finalizado!")
df_final.head()
```

```python
# Ordenando do estado com maior letalidade para o menor em Janeiro
ranking_jan = df_final[['Região/Unidade da Federação', 'Letalidade_Jan']].sort_values(by='Letalidade_Jan', ascending=False)

print("Taxa de Letalidade por Insuficiência Cardíaca em Janeiro/2024:")
print(ranking_jan)
```

```python
import matplotlib.pyplot as plt

# 1. Definindo os meses para o eixo X do gráfico
meses_lista = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
colunas_calculadas = [f'Letalidade_{m}' for m in meses_lista]

# 2. Vamos escolher um estado para ver a curva.
# Vou usar o Rio de Janeiro como exemplo, mas você pode trocar o nome!
nome_estado = '.. Rio de Janeiro'
dados_estado = df_final[df_final['Região/Unidade da Federação'] == nome_estado]

# 3. Pegando os valores de letalidade desse estado
valores_letalidade = dados_estado[colunas_calculadas].values.flatten()

# 4. Desenhando o gráfico
plt.figure(figsize=(12, 6))
plt.plot(meses_lista, valores_letalidade, marker='o', color='red', linewidth=2)

# Deixando o gráfico com cara de "Mestrado na Itália"
plt.title(f'Evolução da Letalidade por Insuficiência Cardíaca - {nome_estado} (2024)', fontsize=14)
plt.xlabel('Mês', fontsize=12)
plt.ylabel('Taxa de Letalidade (%)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Configurações de meses
meses_lista = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
colunas = [f'Letalidade_{m}' for m in meses_lista]

plt.figure(figsize=(15, 8))

# 2. Loop para plotar cada estado
for index, row in df_final.iterrows():
    nome = row['Região/Unidade da Federação']
    valores = row[colunas].values

    # Destacando o TOTAL e os que você já analisou
    if nome == 'Total':
        plt.plot(meses_lista, valores, color='black', linewidth=4, label='BRASIL (MÉDIA)', zorder=10)
    elif nome == '.. Rio de Janeiro':
        plt.plot(meses_lista, valores, color='red', linewidth=2, label='Rio de Janeiro', zorder=5)
    elif nome == '.. São Paulo':
        plt.plot(meses_lista, valores, color='blue', linewidth=2, label='São Paulo', zorder=5)
    else:
        # Os outros estados ficam em cinza para não poluir
        plt.plot(meses_lista, valores, color='gray', alpha=0.3, linewidth=1)

# 3. Ajustes de estética
plt.title('Letalidade por Insuficiência Cardíaca em todos os Estados (2024)', fontsize=16)
plt.ylabel('Taxa de Letalidade (%)', fontsize=12)
plt.xlabel('Meses', fontsize=12)
plt.legend(loc='upper right')
plt.grid(True, alpha=0.2)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados que você obteve na análise
estados = [
    '.. Alagoas',
    '.. Rio de Janeiro',
    '.. Acre',
    '.. Paraíba',
    '.. Amapá'
]
valores = [3.26, 3.01, 2.95, 2.28, 2.22]

# Criando a figura
fig, ax = plt.subplots(figsize=(8, 4))
ax.axis('off') # Esconde os eixos do gráfico

# Criando a tabela
# O formatador '{:.2f}%' garante as duas casas decimais e o símbolo de porcentagem
dados_tabela = [[e.replace('.. ', ''), f'{v:.2f}%'] for e, v in zip(estados, valores)]
colunas = ['Estado (UF)', 'Letalidade Média (2024)']

tabela = ax.table(cellText=dados_tabela,
                  colLabels=colunas,
                  loc='center',
                  cellLoc='center')

# Estilização da Tabela
tabela.auto_set_font_size(False)
tabela.set_fontsize(12)
tabela.scale(1.2, 2) # Ajusta o tamanho das células

# Estilizando o cabeçalho
for (row, col), cell in tabela.get_celld().items():
    if row == 0:
        cell.set_text_props(weight='bold', color='black')
        cell.set_facecolor('#0000') # Azul escuro médico
    else:
        cell.set_facecolor('#0000') # Fundo clarinho para as linhas

plt.title('Top 5: Maior Letalidade por Insuficiência Cardíaca no Brasil',
          fontsize=14, pad=20, weight='bold', color='#e63946')

plt.figtext(0.5, 0.1, 'Fonte: DATASUS / Painel SIH-SUS (Processado via Python)',
            ha='center', fontsize=10, style='italic')

# Comando para salvar a imagem
plt.savefig('tabela_letalidade_top5.png', bbox_inches='tight', dpi=300)
plt.show()
```
